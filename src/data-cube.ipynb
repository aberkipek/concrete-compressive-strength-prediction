# 3-D Data Cube Visualization
Here, we visualized our dataset with using a data cube structure. We plotted 3-D cubes for both initial and cleaned-preprocessed data. We used different markers and colors for making it visually more appealing. We chose 3 features/dimensions according to the corresponding column names in our dateset; Cement, Water and Age, respectively.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler

file_path = "../data/concrete_data.xls"
data = pd.read_excel(file_path)

data.replace(0, data.mean(), inplace=True)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

scaled_df = pd.DataFrame(scaled_data, columns=data.columns)

# Our 3 features for the 3D plot after scaling
x_scaled = scaled_df['Cement (component 1)(kg in a m^3 mixture)']
y_scaled = scaled_df['Water  (component 4)(kg in a m^3 mixture)']
z_scaled = scaled_df['Age (day)']

# Our 3 features for the 3D plot without scaling (original data)
x_original = data['Cement (component 1)(kg in a m^3 mixture)']
y_original = data['Water  (component 4)(kg in a m^3 mixture)']
z_original = data['Age (day)']

fig = plt.figure(figsize=(12, 12))

# We are plotting our data cube with original data
ax1 = fig.add_subplot(211, projection='3d')
ax1.scatter(x_original, y_original, z_original, c='r', marker='o')
ax1.set_xlabel('Cement')
ax1.set_ylabel('Water')
ax1.set_zlabel('Age', rotation=90)
ax1.set_title('Original Data Cube (Before Preprocessing)')

# We are plotting our data cube with preprocessed data
ax2 = fig.add_subplot(212, projection='3d')
ax2.scatter(x_scaled, y_scaled, z_scaled, c='b', marker='^')
ax2.set_xlabel('Cement')
ax2.set_ylabel('Water')
ax2.set_zlabel('Age', rotation=90)
ax2.set_title('Preprocessed Data Cube (After Scaling)')

# We are adjusting the layout to prevent overlap
plt.tight_layout()

plt.show()
```
